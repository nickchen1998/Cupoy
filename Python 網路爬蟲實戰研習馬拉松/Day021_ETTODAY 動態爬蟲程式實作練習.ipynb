## 作業目標

## 1. 取出今天所有的新聞

```python
'''
Your Code
'''
from selenium import webdriver
from bs4 import BeautifulSoup
import time

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

result_news = []
result_date = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    result_date.append(d.find(class_="date").text)
    result_news.append(d.find_all('a')[-1].text)

browser.close()
```

## 2. 取出現在時間兩小時內的新聞

```python
'''
Your Code
'''
import datetime

# 取得今天日期
today_time = datetime.datetime.now()

for i in range(len(result_news)):
    # 去除多餘空白並轉換為時間物件
    tmp = result_date[i].split(" ")
    tmp_time = datetime.datetime.strptime(tmp[0]+" "+tmp[1], '%Y/%m/%d %H:%M')
    
    # 判斷
    if int(str(today_time-tmp_time)[:1]) < 2:
        print(result_date[i]+" "+result_news[i])
        
    # 不讓他跑太多行出來 免得 github 當機
    if i == 20:
        break
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
'''
Your Code
'''
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time ,requests, datetime, random

# 先利用 requests 找出選擇日期的下拉是選單
url = 'https://www.ettoday.net/news/news-list.htm'
result = requests.get(url)

soup = BeautifulSoup(result.text)
print(soup.find(class_='part_calendar_1'))
```

```python
# 取得今天日期
today_time = datetime.datetime.now()

# 計算三天前日期
tmp = today_time - datetime.timedelta(3)
time_list = str(tmp)[:11].split('-')

print(time_list)
```

```python
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

# 設定下拉式選單物件
select_year = Select(browser.find_element_by_name('selY'))
select_mounth = Select(browser.find_element_by_name('selM'))
select_day = Select(browser.find_element_by_name('selD'))

# 設定要選擇的日期
select_year.select_by_value(time_list[0])
select_mounth.select_by_value(str(int(time_list[1])))
select_day.select_by_value(str(int(time_list[2])))


# 設定查詢按鈕物件並執行點擊動作
search_btn = browser.find_element_by_id('button')
search_btn.click()
#等待網頁載入完畢
time.sleep(5)

delay_choices = [8, 5, 10, 6, 20, 11]  #延遲的秒數
delay = random.choice(delay_choices)

# 擷取網頁內容
for i in range(50):
    time.sleep(delay)
    # 此 script 為滑動到底部
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

result_news = []
result_date = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    result_date.append(d.find(class_="date").text)
    result_news.append(d.find_all('a')[-1].text)

browser.close()
print(len(result_news))
```

```python
for i in range(len(result_news)):
    # 去除多餘空白並轉換為時間物件
    tmp = result_date[i].split(" ")
    hour_split = tmp[1].split(":")
    
    # 判斷
    if 15 <= int(hour_split[0]) <= 17:
        print(result_date[i], result_news[i])
```
