### 2. 剛剛的做法只能取出第一頁，試著取出所有頁面的資料。

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
```

```python
# 1. 先取出全部有幾頁
    
'''
Your Code
'''
print(soup.find_all(class_='rc-pagination-item')[-1].text)
```

```python
# 2. 每頁都取出資料放入 courses
import time, random

courses = []

delay_choices = [8, 5, 10, 6, 20, 11]  #延遲的秒數

'''Your Code'''
# 先抓第一頁的資料
for i in soup.find_all('h4'):
    courses.append(i.text)

# 接著選定下一頁按鈕並點擊
for i in range(int(soup.find_all(class_='rc-pagination-item')[-1].text)-1):
    '''
    Your Code
    '''
    try:
        search_btn = browser.find_element_by_class_name('rc-pagination-next')
        search_btn.click()
        time.sleep(random.choice(delay_choices))

        html_source = browser.page_source
        soup = BeautifulSoup(html_source, 'html.parser')

        for j in soup.find_all('h4'):
            courses.append(j.text)
    except:
        print('找不到下一頁按鈕')
        continue
```

```python
# 3. done!
browser.close()
print(courses)
```
