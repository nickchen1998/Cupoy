```python
import json
import requests
from pprint import pprint
```

## 作業一: 練習使用一個民生物聯網 API，例如空氣、地震等感測站所回傳的紀錄資料。

```python
url = 'https://sta.ci.taiwan.gov.tw/STA_Earthquake_v2/v1.0/Things?$expand=Locations&$filter=properties/authority%20eq%20%27%E5%9C%8B%E9%9C%87%E4%B8%AD%E5%BF%83%27%20and%20substringof(%27%E5%9C%B0%E9%9C%87%E7%9B%A3%E6%B8%AC%E7%AB%99%27,name)&$count=true'

resp = requests.get(url)
result = json.loads(resp.text)
pprint([result['value'][index]for index in range(5)])
```

```python
for i in range(10):
    print("測站縣市: " + result['value'][i]['properties']['city'])
    print("隸屬單位: " + result['value'][i]['properties']['authority'])
    print("測站名稱: " + result['value'][i]['properties']['stationName'])
    print("測站編號: " + result['value'][i]['properties']['stationID'])
    print()
```

## 作業二: 練習操作 OGC SensorThings API，將環保局測站位置的資料抓取下來，並且觀察下載資料的內容。

```python
url = 'https://sta.ci.taiwan.gov.tw/STA_AirQuality_EPAIoT/v1.0/Things'

resp = requests.get(url)
result = json.loads(resp.text)

pprint([result['value'][index]for index in range(5)])
```

```python
for i in range(5):
    print("管轄單位: " + result['value'][i]['properties']['authority'])
    print("所在縣市: " + result['value'][i]['properties']['city'] + " " + result['value'][i]['properties']['areaDescription'])
    print("區域型態: " + result['value'][i]['properties']['areaType'])
    print("測站簡述: " + result['value'][i]['properties']['authority'])
    print("感測器網址: " + result['value'][i]['Datastreams@iot.navigationLink'])
    
    # 感測器資料集網址
    url = result['value'][i]['Datastreams@iot.navigationLink']
    datas = requests.get(url).json()
    
    for j in range(len(datas['value'])):
        if datas['value'][j]['description'] == '細懸浮微粒 PM2.5':
            print(datas['value'][j]['description'])
            url = datas['value'][j]['Observations@iot.navigationLink']
            break
    
    # 細懸浮微粒 PM2.5資料網址
    sensor_datas = requests.get(url).json()
    
    for x in range(len(sensor_datas['value'])):
        print("觀測時間: " + sensor_datas['value'][x]['phenomenonTime'])
        print("觀測數值: {}".format(sensor_datas['value'][x]['result']))
        
        if x >= 3:
            break
    
   
    
    print()
```
